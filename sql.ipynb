```python
import pandas as pd 
import sqlite3
import glob
```

### Get List of available SQL Database

```python
database_lst = glob.glob("./db/*.db")
database_lst.sort()
database_lst
```

### Connect to Adventureworks Database

```python
conn = sqlite3.connect(database_lst[0])
```

### List all the tables

```python
tables = list(pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", conn)['name'])
tables
```

### Run SQL Query on the Database

```python
sql_query = 'select * from Products'
df = pd.read_sql_query(sql_query, conn)
df
```
